## Tree implementation using List of List

```
def BinaryTree(r):
  '''
  Design a binary Tree
  '''
  return [r,[],[]] #return root and empty left and right node
```

```
def insertLeft(root,newbranch):
  '''
  insert value into the left node
  '''
  l = root.pop(1)
  print("value of t is",l)
  if len(l) > 1:
    print("i'm inside if because len of l is ", len(l))
    root.insert(1,[newbranch,l,[]])
  else:
    print("i'm inside else because len of l is ", len(l))
    root.insert(1,[newbranch,[],[]])
  return root
```

```
def insertRight(root,newbranch):
  '''
  insert value into the right node
  '''
  t = root.pop(2)
  print("value of t is",t)
  if len(t) > 1:
    print("i'm inside if because len of t is ", len(t))
    root.insert(2,[newbranch,[],t])
  else:
    print("i'm inside else because len of t is ", len(t))
    root.insert(2,[newbranch,[],[]])
  return root
```

```
def setRootVal(root, newVal):
  '''
  setting the root node value
  '''
  root[0] = newVal

def getRootVal(root):
  '''
  fetching the root node value
  '''
  return root[0]

def getLeftChild(root):
  '''
  get the value of the left node
  '''
  return root[1]

def getRightChild(root):
  '''
  get the value of the left node
  '''
  return root[2]
```

```
r = BinaryTree(3)
```

```
insertLeft(r,4)
insertRight(r,10)
```

```
insertRight(r,5)
setRootVal(r,6)
insertRight(r,2)
```

# Tree implementation using Node and References

```
class BT(object):
  '''
  Implement Binary Tree using Node and References
  '''
  def __init__(self,root):
    '''
    Initialize the root object using constractor
    '''
    self.key = root
    self.leftChild = None
    self.rightChild = None
  
  def insertLeftNode(self,newNode):
    '''
    Insert value into left branch
    '''
    if self.leftChild == None:
      self.leftChild = BT(newNode)
    else:
      t = BT(newNode)
      t.leftChild = self.leftChild
      self.leftChild = t

  def insertRightNode(self, newNode):
      '''
      Insert value into right branch
      '''
      if self.rightChild == None:
        self.rightChild = BT(newNode)
      else:
        t = BT(newNode)
        t.rightChild = self.insertRightNode
        self.rightChild = t

  def getLeftChildNode(self):
      return self.leftChild

  def getRightChildNode(self):
      return self.rightChild
    
  def getRootNode(self):
      return self.key

  def setRootNode(self,val):
      self.key = val

```

```
r = BT(3)
r.insertLeftNode(4)
r.insertRightNode(10)
print(r.getRootNode())
print(r.getLeftChildNode().getRootNode())
print(r.getRightChildNode().getRootNode())
```
