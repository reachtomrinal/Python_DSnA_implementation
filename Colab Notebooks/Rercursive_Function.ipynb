# reverse Of a String

```
def reverseStr(str):
  d = {}
  if len(str) <= 1:
    return str
  else:
    if str not in d:
      d[str] = reverseStr(str[1:])
  return d[str] + str[0]
```

```
reverseStr('abc')
```

# Permutation and Combination

```
def perm(s):
  out =[]
  if len(s) == 1:
    out = [s]
    print(f"out value is {out}")
    return out
  else:
    for i,let in enumerate(s): # add a counter and return iterable
      print(f"i is {i} and let is {let}")
      for p in perm(s[:i] + s[i+1:]):
        print(f"value of p is {p}")
        out += [let + p]
        print(out)
    return out
```

```
perm("abc")
```
