# Balanced paranthesis check

```
def balancedParam(s):
  if len(s)%2 != 0:
    return False

  openparam = set('({[')
  matching = set([('(',')'),('{','}'),('[',']')])
  p = []

  for i in s:
    if i in openparam:
      p.append(i)
    else:
      if len(p) == 0:
        return False
      last_open = p.pop()
      if (last_open,i) not in matching:
        return False
  return len(p) == 0  
```
